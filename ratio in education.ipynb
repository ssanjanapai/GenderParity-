```python
import plotly.express as px
import pandas as pd

# Choose the countries you want to include
selected_countries = ['Afghanistan']  # Add more countries as needed

# Filter the DataFrame for selected countries
selected_data = df[df['country'].isin(selected_countries)]

# Create an interactive line plot using Plotly Express
fig = px.line(selected_data.melt(id_vars=['country'], var_name='Year', value_name='Girl/Boy Ratio'),
              x='Year', y='Girl/Boy Ratio', color='country', line_group='country',
              markers=True, line_dash='country', title='Girl/Boy Ratio in Primary and Secondary Education Over Years',
              labels={'Girl/Boy Ratio': 'Girl/Boy Ratio', 'Year': 'Year', 'country': 'Country'},
              template='plotly_dark')  # Use a dark template for a visually appealing look

# Customize the layout
fig.update_layout(xaxis_title='Year', yaxis_title='Girl/Boy Ratio',
                  legend=dict(title='Country', orientation='h', y=1, x=0.5),
                  hovermode='x', hoverlabel=dict(bgcolor='white', font_size=12),
                  plot_bgcolor='#232323', paper_bgcolor='#232323')  # Set background colors

# Show the interactive plot
fig.show()
```

```python
import plotly.express as px

# Create an interactive line plot using Plotly Express
fig = px.line(df.melt(id_vars=['country'], var_name='Year', value_name='Girl/Boy Ratio'),
              x='Year', y='Girl/Boy Ratio', color='country',
              labels={'Girl/Boy Ratio': 'Girl/Boy Ratio', 'Year': 'Year', 'country': 'Country'},
              template='plotly_dark')  # Use a dark template for a visually appealing look

# Customize the layout
fig.update_layout(title='Girl/Boy Ratio in Primary and Secondary Education Over Years',
                  xaxis_title='Year', yaxis_title='Girl/Boy Ratio',
                  legend=dict(title='Country', orientation='h', yanchor='bottom', xanchor='center', y=1.02, x=0.5),
                  hovermode='x unified',
                  plot_bgcolor='#232323', paper_bgcolor='#232323', font_color='white')  # Set background colors

# Create a dropdown widget
country_dropdown = [dict(label=country, method='update',
                         args=[{'visible': df['country'] == country},
                               {'title': f'Girl/Boy Ratio in Primary and Secondary Education Over Years - {country}'}])
                    for country in df['country'].unique()]

# Set up the layout for the dropdown menu
fig.update_layout(updatemenus=[{'type': 'dropdown', 'x': 0.1, 'y': 2,
                                'buttons': country_dropdown,
                                'showactive': True,
                                'active': 0}])

# Show the interactive plot
fig.show()
```

```python
# Pivot the dataframe for creating a heatmap
import matplotlib.pyplot as plt
heatmap_data = df.melt(id_vars=['country'], var_name='Year', value_name='Girl/Boy Ratio')
heatmap_data['Year'] = heatmap_data['Year'].astype(int)

# Set up the matplotlib figure
plt.figure(figsize=(16, 12))

# Plot the data
sns.heatmap(heatmap_data.pivot_table(index='country', columns='Year', values='Girl/Boy Ratio'), cmap='YlGnBu')

# Customize the plot
plt.title('Girl/Boy Ratio in Primary and Secondary Education Over Years (Heatmap)')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```
